## Ejercicio I

```python
def contar_filas(name):
  with open(name) as data:
    contador = 0

    for line in data:
      movie = line.strip().split("::")
      contador += 1

  print(f"La cantidad de filas del archivo es de: {contador}")
```

```python
contar_filas("movies.txt")
```

## Ejercicio II

```python
def contar_c_especiales(name):
  with open(name) as data:
    c_especiales = 0

    for line in data:
      movie = line.strip().split("::")
      for i in line:
        if not i.isalnum():
          c_especiales += 1

  print(f"La cantidad de caracteres especiales es de: {c_especiales}")
```

```python
contar_c_especiales("movies.txt")
```

## Ejercicio III

```python
def contar_peliculas(name):
  with open(name) as data:
    year_movies = {}
    year_movies["Sin año"] = 0

    for line in data:
      movie = line.strip().split("::")
      year = movie[1][-5:-1]
      if year.isnumeric():
        year_movies[year] = year_movies.get(year, 0) + 1
      else:
        year_movies["Sin año"] += 1

  print(year_movies)
```

```python
contar_peliculas("movies.txt")
```

## Ejercicio IV

```python
def largo_nombres():
  with open("movies.txt") as data:
    first_line = data.readline().strip().split("::")
    corta = first_line[1][0:-7]
    larga = first_line[1][0:-7]

    for line in data:
      movie = line.strip().split("::")
      if movie[1][-6] == "(" and movie[1][-1] == ")":
        name = movie[1][:-7]
        if len(name) > len(larga):
          larga = name
        elif len(name) < len(corta):
          corta = name

    print(f'El nombre de la película más larga es: "{larga}", con {len(larga)} caracteres.')
    print(f'El nombre de la película más corta es: "{corta}", con {len(corta)} caracteres.')
```

```python
largo_nombres()
```

## Ejercicio V

```python
def full_gender():
  with open("movies.txt") as data:
    generos_sin_repetir = set()
    for line in data:
      movie = line.strip().split("::")
      generos = movie[-1].split("|")
      for genero in generos:
        generos_sin_repetir.add(genero)

  print(f"Los generos sin repetir son: {generos_sin_repetir}")
```

```python
full_gender()
```

## Ejercicio VI

```python
def min_3_gender():
  with open("movies.txt") as data:
    movies = []
    for line in data:
      movie = line.strip().split("::")
      generos = movie[-1].split("|")
      if len(generos) >= 3:
        movies.append(movie[1])

  print(f"Las películas con mínimo 3 géneros son: {movies}")
```

```python
min_3_gender()
```

## Ejercicio VII

```python
def movies_letters():
  with open("movies.txt") as data:
    generos = {}
    for line in data:
      movie = line.strip().split("::")
      genero = movie[-1].split("|")
      for i in range(len(genero)):
        generos[genero[i].upper()] = len(genero[i])

  print(f"El diccionario con los géneros es: {generos}")
```

```python
movies_letters()
```

## Ejercicio VIII

```python
def movies_filtered():
  with open("movies.txt") as data:
    movies = []
    for line in data:
      movie = line.strip().split("::")
      peliculas_un_genero = lambda movie : movies.append(movie[1]) if "|" not in movie[-1] else None
      peliculas_un_genero(movie)

  print(f"Las películas con sólo un género ({len(movies)}) son: {movies}")
```

```python
movies_filtered()
```

## Ejercicio IX

```python
def map():
  with open("movies.txt") as data:
    movies = {}
    for line in data:
      movie = line.strip().split("::")
      generos = movie[-1].split("|")
      filtro = lambda movie, generos : movies[movie[1]] = generos[0] if "|" not in movie[-1]
      filtro(movie)
```

```python
map()
```
